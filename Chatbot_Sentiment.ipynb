```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

conversation_history = []   # Stores all user messages

def get_sentiment(text):
    score = analyzer.polarity_scores(text)

    if score['compound'] >= 0.05:
        return "Positive"
    elif score['compound'] <= -0.05:
        return "Negative"
    else:
        return "Neutral"

print("Chatbot: Hello! Type 'bye' to end the chat.\n")

while True:
    user_msg = input("You: ")

    if user_msg.lower() == "bye":
        print("\nChatbot: Ending conversation...\n")
        break

    # Save message into history
    conversation_history.append(user_msg)

    # Tier 2: Message-level sentiment
    sentiment = get_sentiment(user_msg)
    print(f"→ Sentiment: {sentiment}")

    # Simple chatbot reply
    print("Chatbot: I understand. Tell me more.\n")

# ---- TIER 1: Conversation-level sentiment ----

print("----- FINAL OUTPUT -----")

overall_scores = [analyzer.polarity_scores(msg)['compound'] for msg in conversation_history]
final_score = sum(overall_scores) / len(overall_scores)

if final_score >= 0.05:
    overall_sentiment = "Positive"
elif final_score <= -0.05:
    overall_sentiment = "Negative"
else:
    overall_sentiment = "Neutral"

print(f"Overall conversation sentiment: {overall_sentiment}")
```
